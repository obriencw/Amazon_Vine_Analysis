```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
vine_data_to_load = "vine_table.csv"
```

```python
# Read the school data file and store it in a Pandas DataFrame.
vine_data_df = pd.read_csv(vine_data_to_load)
vine_data_df.head()
```

```python
vine_data_df.dtypes
```

```python
# Filter on total votes >= 20
total_votes_df = vine_data_df[vine_data_df["total_votes"] >= 20.0]
total_votes_df
```

```python
# Retrieve all rows where helpful_votes divded by total_votes is equal to or greater than 50%
helpful_votes_df = total_votes_df[total_votes_df["helpful_votes"]/total_votes_df["total_votes"] >= .5]
helpful_votes_df
```

```python
# Retrieve all rows where a review was written as part of the Vine program
vine_y_df = helpful_votes_df[helpful_votes_df["vine"] == "Y"]
vine_y_df
```

```python
# Retrieve all rows where a review was not written as part of the Vine program
vine_n_df = helpful_votes_df[helpful_votes_df["vine"] == "N"]
vine_n_df
```

```python
# Determine the total number of reviews
review_count = vine_data_df.count()[0]
print(f'The total number of reviews is {review_count}')
```

```python
# Determine the number of 5 star reviews
vine_five_star = vine_data_df[vine_data_df["star_rating"] == 5]
five_star_count = vine_five_star.count()[0]
print(f'The number of 5-star reviews is {five_star_count}')
```

```python
# Determine the percentage of 5-star reviews for paid
vine_five_star_paid = vine_y_df[vine_y_df["star_rating"] == 5]
five_star_percentage_paid = vine_five_star_paid.count()/vine_y_df.count()[0]
print(f'The percentage of 5-star reviews for paid Vine members is {five_star_percentage_paid[0] * 100:.2f}%')
```

```python
# Determine the percentage of 5-star reviews for unpaid
vine_five_star_unpaid = vine_n_df[vine_n_df["star_rating"] == 5]
five_star_percentage_unpaid = vine_five_star_unpaid.count()/vine_n_df.count()[0]
print(f'The percentage of 5-star reviews for unpaid Vine members is {five_star_percentage_unpaid[0] * 100:.2f}%')
```
